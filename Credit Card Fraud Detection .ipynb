***IMPORTING LIBRARIES***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

***IMPORTING DATASET***

```python
df=pd.read_csv("C:\\Users\\Smita Akolkar\\OneDrive\\Attachments\\Desktop\\Project Data\\creditcard.csv")
df.head()
```

***DATA DESCRIPTION***

```python
df.info()
```

There is no null value in the dataset

```python
df.describe().T
```

```python
##Fixing the values of x and y
x=df.drop(["Class"],axis="columns")

y=df["Class"]


###Spliting data into 70% and 30%
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

***LOGISTRIC REGRESSION***

```python
model=linear_model.LogisticRegression()
model.fit(x_train,y_train)
```

```python
##Accuracy of the logistic model

accuracy=model.score(x_test,y_test)

print("Accuracy of the logistic model is ;",accuracy)
```

***Random Forest Classifier***

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
##Accuracy of the Random forest classifier

accuracy=rf.score(x_test,y_test)

print("Accuracy of the Random Forest Classifier is :",accuracy)
```

***Decision Tree***

```python
from sklearn import tree
tree=tree.DecisionTreeClassifier()
tree.fit(x_train,y_train)
```

```python
###Accuracy of the decision tree classifier

accuracy=tree.score(x_test,y_test)

print("Accuracy of the decision tree classifier is :",accuracy)
```

***KNN***

```python
knn=KNeighborsClassifier()

knn.fit(x_train,y_train)
```

```python
##Accuracy of the KNN

accuracy=knn.score(x_test,y_test)

y_pred=knn.predict(x_test)

print("Accuracy of the KNN :",accuracy)
```

```python
from sklearn.metrics import confusion_matrix,classification_report

conf_matrix=confusion_matrix(y_test,y_pred)

print("\nconfusion matrix\n",conf_matrix)


##Classification report

class_report=classification_report(y_test,y_pred)

print("\nClassification Report\n",class_report)
```
